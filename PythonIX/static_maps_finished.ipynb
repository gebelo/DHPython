# Python2Maps: Matplotlib and Basemap
You need a static map to handle out to the graphics department of your newspaper? Then you might look at Matplotlib in combination with the library Basemap. This tutorial helps you during the very first steps. One way to produce static SVG-maps. 

```python
# Import some very useful libraries. 
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Do some Jupyter Notebook magic to display the output inline. 
%matplotlib inline
```

To get the appropriate bounding box, visit [boundingbox.klokantech.com](http://boundingbox.klokantech.com). Chose 'CSV' format in the box at the bottom of the page. 

```python
# The collegues want you to make a map. Ok, let's go. 

# Set the area. 
area = [2.3889,49.497,6.4081,51.5517] # Belgium

# Open a canvas. 
fig = plt.figure(figsize=[15,10])

# Create a Basemap object setting the coordinates around the box surrounding area.
m = Basemap(llcrnrlon=area[0], llcrnrlat=area[1], urcrnrlon=area[2], urcrnrlat=area[3], resolution='i', epsg=3857)

# Draw boundaries, coastlines and map background
m.drawcoastlines()
m.drawcountries()

# Show the map. 
m
```

```python
# Well -- at least we got something. We need WATER! 

# Set the area. 
area = [2.3889,49.497,6.4081,51.5517] # Belgium

# Open a canvas. 
fig = plt.figure(figsize=[15,10])

# Create a Basemap object setting the coordinates around the box surrounding area.
m = Basemap(llcrnrlon=area[0], llcrnrlat=area[1], urcrnrlon=area[2], urcrnrlat=area[3], resolution='i', epsg=3857)

# Draw boundaries, coastlines and map background
m.drawcoastlines()
m.fillcontinents()
m.drawcountries()
m.drawrivers()

# Show the map. 
m
```

It works for the US. But it doesn't for Europe. Well then, let's go on. We won't compain. And we will add 
the districts manually. First, we need a shapefile with the districts of Belgium. Working with shapefiles can be nasty. If you are getting in trouble, first check the [Basemap documentation](https://basemaptutorial.readthedocs.io/en/latest/shapefile.html).

I downloaded the shapefile for Belgium from [GADM](https://gadm.org/download_country_v3.html). If you are working with other shapefiles, you maybe need to change the projection to EPSG:4326. You can do this for example with QGis or Ogr2Ogr.

```python
# It works for the US. But it doesn't for Europe. Well then, let's go on. We won't compain. And we will add 
# the Provinces manually.  

# Set the area. 
area = [2.3889,49.497,6.4081,51.5517] # Belgium

# Open a canvas. 
fig = plt.figure(figsize=[15,10])

# Create a Basemap object setting the coordinates around the box surrounding area.
m = Basemap(llcrnrlon=area[0], llcrnrlat=area[1], urcrnrlon=area[2], urcrnrlat=area[3], resolution='i', epsg=3857)

# Draw boundaries, coastlines and map background
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='beige', lake_color='aqua')
m.drawcoastlines()
m.drawrivers(color='aqua')
m.drawcountries()
m.drawstates(zorder=3)

# Reading in the shapefile. 
m.readshapefile('data/shapefiles/gadm36_BEL_2', 'NAME_2')

# Show the map. 
m
```

```python
# Let's put some city names on the map. Let's render the map in full resolution. (It will take long...)
# Finally, we're going to save it. 

# Set the area. 
area = [2.3889,49.497,6.4081,51.5517] # Belgium

# Open a canvas. 
fig = plt.figure(figsize=[15,10])
ax = fig.add_subplot(111, frame_on=False)

# Create a Basemap object setting the coordinates around the box surrounding area.
m = Basemap(llcrnrlon=area[0], llcrnrlat=area[1], urcrnrlon=area[2], urcrnrlat=area[3], resolution='f', epsg=3857)

# Draw boundaries, coastlines and map background
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='beige', lake_color='aqua')
m.drawcoastlines()
m.drawrivers(color='aqua')
m.drawcountries()
m.drawstates(zorder=3)

# Reading in the shapefile. 
m.readshapefile('data/shapefiles/gadm36_BEL_2', 'NAME_2')

# Add some labels for the most important cities. 
cities = {'Antwerp': [51.21989, 4.40346], 'Ghent' : [51.05, 3.71667], 
          'Charleroi' : [50.4167,4.4333], 'Liège' : [50.63373, 5.56749], 
          'Bruxelles' : [50.833,4.333]}

for c in cities: 
    m.scatter(cities[c][1], cities[c][0], latlon=True, color='k')
   
    # Convert from latitude/longitude to x,y
    x, y = m(cities[c][1],cities[c][0])
    ax.annotate(c, xy=(x, y), xycoords='data', xytext=(x, y), textcoords='data')

# Save the map
plt.savefig('output/amazingmap.svg', bbox_inches='tight')

# Show the map. 
m
```
